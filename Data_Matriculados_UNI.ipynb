## *Nombre del Dataset:	Alumnos matriculados en la Universidad Nacional de Ingeniería - [UNI]*

*Listado de alumnos matriculador en la Universidad Nacional de Ingeniería desde 2016-1 al 2024-1.*

**IDHASH:** Identificador anónimo único para cada alumno, si el alumno participa en más de un período de admisión debería aparecer nuevamente con el mismo identificador.

**COLEGIO_DEPA:** Departamento donde se encuentra ubicado el colegio del alumno.

**COLEGIO_PROV:** Provincia donde se encuentra ubicado el colegio del alumno.

**COLEGIO_DIST:** Distrito donde se encuentra ubicado el colegio del alumno.

**ANIO:** Año en que se matricula el alumno.

**PERIODO:** Ciclo del año en que matricula el alumno.

**TIPO_MATRICULA:** Tipo de matrícula para el alumno: Regulares, Cachimbos.

**DOMICILIO_DEPA:** Departamento donde se encuentra ubicado el domicilio del alumno.

**DOMICILIO_PROV:** Provincia donde se encuentra ubicado el domicilio del alumno.

**DOMICILIO_DIST:** Distrito donde se encuentra ubicado el domicilio del alumno.

**ANIO_NACIMIENTO:** Año de nacimiento del alumno.

**NACIMIENTO_PAIS:** Nombre del pais de nacimiento del alumno.

**NACIMIENTO_DEPA:** Departamento donde se encuentra ubicado el lugar de nacimiento del alumno.

**NACIMIENTO_PROV:** Provincia donde se encuentra ubicado el lugar de nacimiento del alumno.

**NACIMIENTO_DIST:** Distrito donde se encuentra ubicado el lugar de nacimiento del alumno.

**SEXO:** Sexo del alumno a la universidad: MASCULINO, FEMENINO.

**MODALIDAD:** Modalidad de ingreso: ORDINARIO, INGRESO DIRECTO, GRADUADOS EN LA UNI, DOS PRIMEROS ALUMNOS.

**METODOLOGIA:** Indica si las clases se realizan de manera presencial o virtual.

**FACULTAD:** Nombre de la facultad.

**ESPECIALIDAD:** Nombre de la especialidad profesional a la que pertence el alumno.

**CICLO_RELATIVO:** Avance periódico, su valor es entre 1 y 11.

```python
#Se evidencian errores de correspondencia Especialidad - Facultad
df[df['ESPECIALIDAD'].str.contains('ESTADÍSTICA')]['FACULTAD'].unique()
```

```python
df.loc[df['ESPECIALIDAD'].str.contains('ESTADÍSTICA'), 'FACULTAD'] = 'INGENIERÍA ECONÓMICA, ESTADÍSTICA Y CIENCIAS SOCIALES'
df.loc[df["ESPECIALIDAD"].str.contains('ESTADÍSTICA'),'FACULTAD'].unique()
```

```python
df.loc[df['ESPECIALIDAD'].str.contains('ESTADÍSTICA'), 'ESPECIALIDAD'] = 'INGENIERÍA ESTADÍSTICA'
df.loc[df['ESPECIALIDAD'].str.contains('INGENIERÍA DE SOFTWARE'), 'ESPECIALIDAD'] = 'INGENIERÍA DE SOFTWARE'
df.loc[df["ESPECIALIDAD"]=='INGENIERÍA DE PETRÓLEO', 'ESPECIALIDAD'] = 'INGENIERÍA DE PETRÓLEO Y GAS NATURAL'

df["ESPECIALIDAD"].unique()
```

```python
df['NACIMIENTO_PAIS'].value_counts()
```

```python
df['NACIMIENTO_PAIS'].isna().sum()
```

```python
#Se reemplazan valores faltantes por la moda, además al revisar la data se visualiza que corresponden a Provincias de Perú
df = df.fillna({'NACIMIENTO_PAIS':'Peru'})
df['NACIMIENTO_PAIS'].value_counts().sum()
```

```python
df['TIPO_MATRICULA'].unique()
```

```python
df["MODALIDAD"].value_counts().sort_index()*100/df.shape[0]
```

```python
df['MODALIDAD_MOD'] = df['MODALIDAD'].copy()
df['MODALIDAD_MOD']
```

```python
df.loc[df['MODALIDAD_MOD'].str.contains('CONCURSO NACIONAL ESCOLAR'), 'MODALIDAD_MOD'] = 'CONCURSO NACIONAL ESCOLAR'
df.loc[df['MODALIDAD_MOD'].str.contains('DIPLOMADOS CON BACHILLERATO'), 'MODALIDAD_MOD'] = 'DIPLOMADO CON BACHILLERATO INTERNACIONAL'
df.loc[df['MODALIDAD_MOD'].str.contains('INGRESO DIRECTO'), 'MODALIDAD_MOD'] = 'INGRESO DIRECTO CEPRE-UNI'
df.loc[df['MODALIDAD_MOD']=='TITULADOS O GRADUADOS', 'MODALIDAD_MOD'] = 'TITULADOS O GRADUADOS EN OTRA UNIVERSIDAD'
df.loc[df['MODALIDAD_MOD']=='TITULADOS O GRADUADOS EN LA UNI', 'MODALIDAD_MOD'] = 'TITULADOS O GRADUADOS UNI'
df.loc[(df['MODALIDAD_MOD'].str.contains('VICTIMAS DEL TERRORISMO'))|(df['MODALIDAD_MOD'].str.contains('VÍCTIMAS DEL TERRORISMO')), 'MODALIDAD_MOD'] = 'VICTIMA DEL TERRORISMO'
df['MODALIDAD_MOD'].value_counts().sort_index()
```
